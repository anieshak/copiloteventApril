```python
# Import necessary libraries
import pandas as pd

# Load the dataset
file_path = './data/flights.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Display dataset information
data.info()

# Display summary statistics
data.describe()
```

```python
# Check for null values
null_values = data.isnull().sum()
print("Null values in each column:\n", null_values)

# Replace null values with 0
data.fillna(0, inplace=True)

# Verify there are no null values left
data.isnull().sum()
```

```python
# Group by DayofMonth, OriginAirportID, and DestAirportID
delay_chances = data.groupby(['DayofMonth', 'OriginAirportID', 'DestAirportID'])['DepDel15'].mean()

# Reset index for better readability
delay_chances = delay_chances.reset_index()

# Rename the column for clarity
delay_chances.rename(columns={'DepDel15': 'DelayChance'}, inplace=True)

# Display the first few rows of the result
delay_chances.head()
```

```python
import pickle

# Save the delay_chances DataFrame as a model to a file
model_file_path = './data/delay_chances_model.pkl'
with open(model_file_path, 'wb') as model_file:
    pickle.dump(delay_chances, model_file)
print(f"Model saved to {model_file_path}")

# Extract unique airport IDs and names
airports = data[['OriginAirportID', 'OriginAirportName']].drop_duplicates()

# Save the airports data to a new CSV file
airports_file_path = './data/airports.csv'
airports.to_csv(airports_file_path, index=False)
print(f"Airports data saved to {airports_file_path}")
```
